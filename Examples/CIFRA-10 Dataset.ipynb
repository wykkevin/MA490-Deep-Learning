```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
CIFAR10 = np.load('CIFAR-10_train.npz')
CIFAR10.keys()
```

```python
images = CIFAR10['train_images']
labels = CIFAR10['train_labels']
names = CIFAR10['label_names']
print(images.shape)
print(labels.shape)
print(names)
```

```python
ix = 0
img = images[ix,:,:,:]
plt.imshow(img)
print(names[labels[ix]])
```

```python
plt.figure(figsize=(10,10))
for i in range(10):
    plt.subplot(5,2,i+1)
    ix = np.argwhere(labels==i).flatten()[0]
    print(names[i],ix)
    img = images[ix,:,:,:]
#     plt.figure(figsize=(1,1))
    plt.imshow(img,interpolation=None)
    plt.axis('off')
    plt.show()
```
