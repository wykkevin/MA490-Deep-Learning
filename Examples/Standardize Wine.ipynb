```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("wine_quality_red.csv",sep=";")
df.head(5)
```

```python
features = df.drop("quality",axis=1)
target = df['quality']
```

```python
mu = features.mean()
sd = features.std()
features_s = (features - mu)/sd
features_s.head(5)
```

```python
features_s.describe()
# check std都是1
```

```python
features_s['bias']=1
# bias 不standardize
```

```python
X = features_s.values
y = target.values
```

```python
def RMSE(yh,y):
    return np.sqrt(((yh-y)**2).mean())
```

```python
A = np.matmul(X.T, X)
b = np.matmul(X.T, y)
wopt = np.linalg.solve(A,b)
yh = np.matmul(X,wopt)
print("RMSE=",RMSE(yh,y),' baseline',y.std())
pd.Series(wopt,index=features_s.columns).sort_values()
```
