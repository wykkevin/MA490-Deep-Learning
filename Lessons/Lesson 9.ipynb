```python
import numpy as np
import pandas as pd
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def y1(x1,x2):
    return x1+x1*x2+x2**2;
```

```python
def y2(x1,x2):
    return x1**2+x1*x2+x2**2
```

```python
x1 = np.linspace(-100,100,200)
x2 = np.linspace(-100,100,200)
X, Y = np.meshgrid(x1,x2)
Z1=y1(X,Y)
Z2=y2(X,Y)
```

```python
fig=plt.figure()
ax=plt.axes(projection='3d')
ax.contour3D(X,Y,Z1,100,cmap='binary')
ax.set_xlabel('x1')
ax.set_xlabel('x2')
ax.set_xlabel('f')
```

```python
fig=plt.figure()
ax=plt.axes(projection='3d')
ax.contour3D(X,Y,Z2,100,cmap='binary')
ax.set_xlabel('x1')
ax.set_xlabel('x2')
ax.set_xlabel('f')
```

```python
a = np.linspace(0,1,101)
x1=np.array([-5,5])
x2=np.array([5,-5])
```

```python
x= np.zeros([101,2])
k=0
for a0 in a:
    x[k]=a[k]*x1+(1-a[k])*x2
    k=k+1
```

```python
def f(x):
    return x[0]+x[0]*x[1]+x[1]**2
```

```python
y=np.zeros(101)
k=0
for k in range(101):
    y[k]=f(x[k])
```

```python
plt.plot(a,y)
xx = np.arange(-5.1,5,0.1)
plt.plot(a,y1(xx,xx).tolist())
plt.plot(a,y2(xx,xx).tolist())
plt.show()
```

So, the first one is not convex, and the second one is convex.
