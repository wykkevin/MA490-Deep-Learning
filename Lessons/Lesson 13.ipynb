```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def cross_entropy(p,q):
    return -(p*np.log10(q)).sum()
```

```python
def fsmax(x):
    k=x.max()
    e = np.exp(x-k)
    p = e/e.sum()
    return p
```

```python
def MSE(yh,y):
    return ((yh-y)**2).mean()
```

```python
def RMSE(yh,y):
    return np.sqrt(((yh-y)**2).mean())
```

```python
mse = np.zeros(1000)
ce = np.zeros(1000)
for k in range(1000):
    p = 100*np.random.rand(10)
    q = 100*np.random.rand(10)
    mse[k] = MSE(fsmax(p),fsmax(q))
    ce[k] = cross_entropy(fsmax(p),fsmax(q))
```

```python
plt.scatter(x=mse,y=ce,s=50,alpha=0.1)
plt.xlabel('MSE')
plt.ylabel('CE')
plt.xlim(xmin=mse.min(),xmax=mse.max())
plt.ylim(ymin=ce.min(),ymax=ce.max())
```

```python
# MSE has little variation, but CE shows large variations.
# Most of the MSE are at 0.2, while CE varies from 0 to 40
```
