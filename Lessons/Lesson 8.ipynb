```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.array([67,67,72,69,69,70,67,67,66,68])
y = np.array([128,123,129,143,132,142,112,118,108,119])
```

```python
def MSE(w,b,x,y):
    n = len(x)
    yh = w * x + b
    dy = y - yh
    dy2 = dy * dy
    return dy2.mean()
```

```python
def plot_RMSE(wmin,wmax,bmin,bmax,num_pts,x,y):
    rmse = np.zeros((num_pts,num_pts))
    ww,bb = np.meshgrid(np.linspace(wmin,wmax,num_pts),np.linspace(bmin,bmax,num_pts))
    for i in range(num_pts):
        for j in range(num_pts):
            rmse[i,j] = np.sqrt(MSE(ww[i,j],bb[i,j],x,y))
    plt.contour(ww,bb,rmse,25)
    plt.show()
    return (ww,bb,rmse)
```

```python
(ww,bb,rmse)=plot_RMSE(-5,10,-177,-157,100,x,y)
```

```python
mu_x = x.mean()
sd_x = x.std()
x_s = (x - mu_x)/sd_x
```

```python
mu_y = y.mean()
sd_y = y.std()
y_s = (y - mu_y)/sd_y
```

```python
(ww,bb,rmse)=plot_RMSE(-1,1,-1,1,100,x_s,y_s)
```
